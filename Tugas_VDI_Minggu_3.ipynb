# **7. Visualisasi Data**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#Membaca file
df = pd.read_csv('StudentPerformanceFactors_preprocessed.csv')

# Menghitung korelasi Pearson antara semua kolom numerik
correlation_matrix = df.corr()

# Mengambil korelasi antara semua fitur dan kolom target 'Exam_Score'
target_correlation = correlation_matrix[['Exam_Score']].sort_values(by='Exam_Score', ascending=False)

# Visualisasi korelasi dalam bentuk heatmap
plt.figure(figsize=(8, 6))

# Membuat heatmap dengan seaborn
sns.heatmap(target_correlation, annot=True, cmap='coolwarm', linewidths=0.5)

# Menambahkan judul
plt.title('Korelasi Pearson dengan Atribut Exam_Score', fontsize=16)

# Menampilkan heatmap
plt.show()
```

# **Penjelasan Visualisasi**
- Korelasi mendekati 1 berarti hubungan positif yang kuat.(hubungan positif = saling berbanding lurus = semakin tinggi nilai satu fitur semakin tinggi pula atribut target dan sebaliknya)

- Korelasi mendekati -1 berarti hubungan negatif yang kuat. (hubungan negatif = saling berbanding terbalik = semakin tinggi nilai pada suatu fitur, justru semakin rendah nilai pada atribut target nya. Begitupun sebaliknya)

- Korelasi mendekati 0 berarti tidak ada hubungan linier yang signifikan.
